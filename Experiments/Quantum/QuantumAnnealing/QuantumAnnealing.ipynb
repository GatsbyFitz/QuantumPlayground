### Gate-based vs Annealing Quantum Computing

| Aspect | **Quantum Annealing** | **Gate-Based Quantum Computing** |
| --- | --- | --- |
| **Approach** | Find the lowest-energy (optimal) state | Manipulates qubits with quantum gates to build circuits for general-purpose computation |
| **Primary Use Case** | Specialized for optimization problems (e.g., scheduling, logistics, portfolio optimization) | General-purpose algorithms: cryptography, quantum simulation, search, factoring |
| **Hardware** | D-Wave systems with superconducting flux qubits | IBM, Google, Rigetti, IonQ using superconducting qubits or trapped ions |
| **Robustness** | Less prone to gate-level errors but limited error correction | Highly sensitive to decoherence and gate errors  |
| **Maturity** | More mature for specific use cases | Long-term promise for universal quantum computing |

## What is Quantum Annealing?

Finding the optimum solution to problems by searching over a space and finding the global minimum = the optimal solution to the problem. The benefit of Quantum vs Classical Computing is the ability to tunnel and skip to the next lowest point rather than having to explore every single step/part of the energy topology.

Like before, we craft a Hamiltonian that encodes the optimisation problem, where we can input the state of the system (a configuration of qubits), and the energy value associated with that state is the output.

### A Simple Hamiltonian Working Example

I appreciate that this may seem a bit conceptual, so let's work through a simple example. 

Let's start with the Ising Hamiltonian, 

$$H(\{\sigma\}) = -J \sum_{\langle i,j\rangle} \sigma_i \sigma_j - h \sum_i \sigma_i$$

It describes spins $\sigma_i = \pm 1$ on a lattice, where $J$ controls how neighbors align 
(ferromagnetic if $J>0$, antiferromagnetic if $J<0$), and $h$ is an external magnetic field 
biasing spin orientation.

With a two-spin working example, it simplifies to:

$$
H(\sigma_1, \sigma_2) = -J \sigma_1 \sigma_2 - h(\sigma_1 + \sigma_2)
$$

Both up: $\sigma_1 = \sigma_2 = +1$. 
$$
H = -J - 2h
$$

Both down: $\sigma_1 = \sigma_2 = -1$. 
$$
H = -J + 2h
$$

Mixed: $(\sigma_1, \sigma_2) = (+1,-1)$ or $(-1,+1)$. 
$$
H = +J
$$
(same for both mixed states)

This already shows the roles of $J$ and $h$: 
$J>0$ favors alignment; $h$ biases the total spin toward its sign.
